#**Task 2**
**Unemployment is measured by the unemployment rate which is the number of people
 who are unemployed as a percentage of the total labour force. We have seen a sharp
 increase in the unemployment rate during Covid-19, so analyzing the unemployment rate
 can be a good data science project.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Unemployment in India.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.columns=df.columns.str.strip()
```

```python
df["Date"]=pd.to_datetime(df['Date'])
df['Day']=df["Date"].dt.day
df['Month']=df["Date"].dt.month
df['Year']=df["Date"].dt.year
df.head(5)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df=df.sort_values(by="Region")
df.head(5)
```

```python
plt.figure(figsize=(8,9))
sns.histplot(x=df["Estimated Unemployment Rate (%)"],hue=df["Region"])
```

```python
#plt.figure(figsize=(12,12))
sns.barplot(data=df,x="Month",y="Estimated Unemployment Rate (%)",hue="Month")
```

```python
sns.barplot(data=df,x="Year",y="Estimated Unemployment Rate (%)",hue="Year")
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df,x='Month',y="Estimated Unemployment Rate (%)",hue="Region")
```

```python
plt.figure(figsize=(30,10))
sns.barplot(data=df,x='Day',y="Estimated Unemployment Rate (%)",hue="Region")
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df,x='Year',y="Estimated Unemployment Rate (%)",hue="Region")
```

```python
greater_than_mean=df[df["Estimated Unemployment Rate (%)"] >= df["Estimated Unemployment Rate (%)"].mean()]
less_than_mean=df[df["Estimated Unemployment Rate (%)"] < df["Estimated Unemployment Rate (%)"].mean()]
```

```python
#Regions having employment rate higher tham mean
plt.figure(figsize=(20,10))
sns.barplot(data=greater_than_mean , x='Region',y="Estimated Unemployment Rate (%)")
plt.xticks(rotation=45)
```

```python
#Regions having employment rate less tham mean
plt.figure(figsize=(20,10))
sns.barplot(data=less_than_mean, x='Region',y="Estimated Unemployment Rate (%)")
plt.xticks(rotation=45)
```
