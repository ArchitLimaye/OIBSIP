#**Task 1**
**ris flower has three species; setosa, versicolor, and virginica, which differs according to their
 measurements. Now assume that you have the measurements of the iris flowers according to
 their species, and here your task is to train a machine learning model that can learn from the
 measurements of the iris species and classify them.**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Iris.csv")
df.sample(5)
```

```python
df.info()
```

```python
sns.scatterplot(x=df["SepalLengthCm"],y=df["SepalWidthCm"],hue=df["Species"])
```

```python
df.iloc[:,1:-1]
```

```python
from sklearn.preprocessing import StandardScaler
pre=StandardScaler()
df_Scaled=pre.fit_transform(df.iloc[:,1:-1])
```

```python
x=df_Scaled
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.7,random_state=48)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(pred,y_test)
print("The Accuracy of the Model is ",acc*100)
```
